```python
import pandas as pd



#read in csv file and locate it
url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FY006A/CSV/1.0/en"
df = pd.read_csv(url)
df.tail(8)
```

```python
# get the coluumn list to decide what to drop

headers = df.columns.tolist()
headers
```

```python
drop_col_list = [
    'STATISTIC',
    'Statistic Label',
    'TLIST(A1)',
    'CensusYear',
    'C02199V02655',
    'C02076V03371',
    'C03789V04537',
    'UNIT',
    'Administrative Counties'
]

# Drop the specified columns 
df_cleaned = df.drop(columns=drop_col_list, errors='ignore')
df_cleaned.head()
```

```python
# get ride of the "both sexes" in sex column
df_cleaned = df_cleaned[df_cleaned['Sex'] != 'Both sexes']
df_cleaned.head()
```

```python

# create pivot table for analysis
df_anal = pd.pivot_table(df, 'Sex','VALUE',"Single Year of Age", 'Administrative Counties')
print (df_anal.head(3))
# write out the entire file to local machine
df_anal.to_csv("population_for_analysis.csv")




## df anal error!!!!!!
```

```python
#  Weighted mean age (by sex)

# filter rows needed for analysis
df_filtered = df[
    (df['Sex'].isin(['Male', 'Female'])) &
    (df['Single Year of Age'] != 'All ages')
]

# 2. Convert age to numeric safely
df_filtered['Single Year of Age'] = pd.to_numeric(
    df_filtered['Single Year of Age'], errors='coerce'
)
df_filtered = df_filtered.dropna(subset=['Single Year of Age'])
df_filtered['Single Year of Age'] = df_filtered['Single Year of Age'].astype(int)

# 3. Weighted mean function
def weighted_mean_age(x):
    return (x['Single Year of Age'] * x['VALUE']).sum() / x['VALUE'].sum()

# 4. Weighted mean age by sex (Ireland total)
weighted_mean_by_sex = df_filtered.groupby('Sex').apply(weighted_mean_age)

print(weighted_mean_by_sex)
```

```python
# Write that code that would group the people within 5 years of that age together, into one age group
df_filtered['Single Year of Age'] = pd.to_numeric(df_filtered['Single Year of Age'], errors='coerce')
df_filtered = df_filtered.dropna(subset=['Single Year of Age'])
df_filtered['Single Year of Age'] = df_filtered['Single Year of Age'].astype(int)

# 3. Filter for ages within 5 years of the target age
lower_age = age_to_check - 5  # include 5 years younger
upper_age = age_to_check + 5  # include 5 years older

df_age_group = df_filtered[
    (df_filtered['Sex'].isin(['Male', 'Female'])) &
    (df_filtered['Single Year of Age'] >= lower_age) &
    (df_filtered['Single Year of Age'] <= upper_age)
]

# 4. Pivot table: population by county and sex for this age group
df_age_group_pivot = df_age_group.pivot_table(
    values='VALUE',
    index='Administrative Counties',
    columns='Sex',
    aggfunc='sum',
    fill_value=0
)

# 5. Optional: calculate difference between sexes
df_age_group_pivot['Difference'] = (df_age_group_pivot['Male'] - df_age_group_pivot['Female']).abs()

region_with_biggest_difference = df_age_group_pivot['Difference'].idxmax()
biggest_difference_value = df_age_group_pivot['Difference'].max()

print(f"Age group: {lower_age}-{upper_age}")
print(f"County with biggest male–female population difference in this age group:")
print(f"{region_with_biggest_difference} → {biggest_difference_value} people")


```
