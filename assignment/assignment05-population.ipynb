```python
import pandas as pd



#read in csv file and locate it
url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FY006A/CSV/1.0/en"
df = pd.read_csv(url)
df.tail(8)
```

```python
# get the coluumn list to decide what to drop

headers = df.columns.tolist()
headers
```

```python
drop_col_list = [
    'STATISTIC',
    'Statistic Label',
    'TLIST(A1)',
    'CensusYear',
    'C02199V02655',
    'C02076V03371',
    'C03789V04537',
    'UNIT'
]

# Drop the specified columns but keep 'Sex'
df_cleaned = df.drop(columns=drop_col_list, errors='ignore')
df_cleaned.head()
```

```python
# get ride of the "both sexes" in sex column
df_cleaned = df_cleaned[df_cleaned['Sex'] != 'Both sexes']
df_cleaned.head()
```
