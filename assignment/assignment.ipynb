```python
## Creating the Pie Chart

import pandas as pd 
# import matplotlib library
import matplotlib.pyplot as plt


datadir = "../assignment/"
filename = "people-1000.csv"


# read in the file
df = pd.read_csv(datadir + filename)
# to have a look at the data, will show first 5 rows for example, do a print(df) for all data
# print(df.head())  can mark out later if needed

# To extract the domain from the email addresses
df["domain"] = df["Email"].str.split("@").str[1]

# to check it worked, print email and domain columns only
print(df[["Email", "domain"]].head())

# to count how many of each domain there are
print(df["domain"].value_counts())
# this shows that there are 3 domains in the data set

# to create a pie chart of the domain counts
domain_counts = df["domain"].value_counts()

# to make the pie chart

# this is to create the size of the pie chart
plt.figure(figsize=(7,7))
#this is to create the pie chart,domain count for sizes, labels, autopct to show as a percentage and startangle to start at 90 degrees
# for changing colors you can add colors=colors=['color1','color2','color3']-to make it look nicer
colors = ["#ac37cc", "#186d9e", "#d11db6"]
plt.pie(domain_counts, labels=domain_counts.index, autopct='%1.1f%%', startangle=90,colors=colors)

# to add a title 
plt.title("Email Domain Distribution")
# to make the circle
plt.axis('equal')
# to show it 
plt.show()


```
